```python
import os
from urllib.parse import unquote
```

```python
sqliInput = open("./sqliInput.txt", 'w+')
i = 0
with open("./log_5.21_tagged.txt", 'r') as logfile:
    for line in logfile:
        if i < 10000:
            line=line.split(' ')
            if line[0]=='[sqli]':
                i = i + 1
                #res = line[7].find('/vulnerabilities/sqli/?id=')
                #if res == -1:
                    #print(line[7])
                #print(unquote((line[7])))
                inputLine = unquote((line[7])).lstrip('/vulnerabilities/sqli/?id=')
                sqliInput.write(inputLine + '\n')
                #print(inputLine)
        else:
            break
    print(i)
logfile.close()
sqliInput.close()
```
